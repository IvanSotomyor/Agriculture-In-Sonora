```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_excel('Dummies (1).xlsx')
df.head()
```

```python
df.describe()
```

```python
X = df.drop('PRODTON', axis=1)
y = df['PRODTON']
```

```python
X.shape
```

```python
y.shape
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_scaled
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=30)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train, y_train)
print('Score in test model without PCA = {}\nScore in train model without PCA = {}'.format(model.score(X_test, y_test),model.score(X_train, y_train)))
```

##  L1 and L2 regularization sin PCA

---



```python
#   Lasso Regularization = L1,
from sklearn import linear_model
lasso_reg = linear_model.Lasso(alpha=50, max_iter=1000, tol=0.1)
lasso_reg.fit(X_train, y_train)
```

```python
print('L1 Score in test = {} \n \n L1 Score in train = {}'.format(lasso_reg.score(X_test, y_test), lasso_reg.score(X_train, y_train)))
```

```python
#   Ridge Regularization = L2
from sklearn.linear_model import Ridge
ridge_reg = Ridge(alpha=50, max_iter=1000, tol=0.1)
ridge_reg.fit(X_train, y_train)
```

#  PCA

---



```python
from sklearn.decomposition import PCA
#   Retain 95% of information, whatever features you come on with, I'm happy with that
pca = PCA(0.95)
X_pca = pca.fit_transform(X_scaled)
print('Original columns without PCA = {}\nNew columns with PCA = {}'.format(X_scaled.shape, X_pca.shape))
```

```python
pca.explained_variance_ratio_
```

```python
pca.n_components_
```

```python
X_train, X_test, y_train, y_test =train_test_split(X_pca, y, test_size=0.2, random_state=30)
```

```python
pca_model = LinearRegression()
pca_model.fit(X_train, y_train)
print('Score in test model with PCA = {}\nScore in train model with PCA = {}'.format(pca_model.score(X_test, y_test),pca_model.score(X_train, y_train)))
```
